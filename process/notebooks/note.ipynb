```python
import pandas as pd
```

```python
INPUT_FILE = "../data/locations_updated.csv"
```

```python
def get_remaining(df):
   return df[(df["latitude"].isna() | df["longitude"].isna())]

def get_remaining_count(df):
    return len(get_remaining(df))
```

```python
df = pd.read_csv(INPUT_FILE)
```

```python
remaining = get_remaining(df)
```

```python
remaining.to_csv("../data/remaining.csv", index=False)
```

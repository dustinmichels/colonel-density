```python
import pandas as pd
```

```python
INPUT_FILE = "../data/locations.csv"
UPDATED_FILE = "../data/locations_updated.csv"
```

```python
def get_remaining_count(df):
    return len(df[df["latitude"].isna() | df["longitude"].isna()])
```

```python
df = pd.read_csv(INPUT_FILE)
df2 = pd.read_csv(UPDATED_FILE)
```

```python
get_remaining_count(df)
```

```python
get_remaining_count(df2)
```
